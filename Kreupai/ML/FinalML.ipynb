```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("results.csv")
dff=pd.read_csv("matchids.csv")
teamids=pd.read_csv("teamids.csv")
```

```python
merged_df = pd.merge(
    df,
    dff[['matchname', 'season', 'datetime','matchid']], 
    on=['matchname', 'season'], 
    how='left'
)
merged_df = merged_df.merge(
    teamids.rename(columns={"teamName": "team", "teamId": "teamid"}),
    on="team",
    how="left"
)

# Merge for opponent -> opponentid
merged_df = merged_df.merge(
    teamids.rename(columns={"teamName": "opponent", "teamId": "opponentid"}),
    on="opponent",
    how="left"
)
merged_df = merged_df.sort_values(["team", "datetime"]).reset_index(drop=True)
merged_df["team_avg_goals_last5"] = (
    merged_df.groupby("team")["goals_scored"]
    .transform(lambda x: x.shift().rolling(5, min_periods=1).mean())
)

merged_df["team_avg_conceded_last5"] = (
    merged_df.groupby("team")["goals_conceded"]
    .transform(lambda x: x.shift().rolling(5, min_periods=1).mean())
)

# -------------------
# Create a grouping column: teamid if home, opponentid if away
merged_df["group_id"] = np.where(
    merged_df["ishome"] == 1,
    merged_df["team"],
    merged_df["opponent"]
)

# Team's own goals average
merged_df["team_home_goals_avg"] = (
    merged_df.groupby("group_id")["goals_scored"]
    .expanding()
    .mean()
    .reset_index(level=0, drop=True)
)







points_map = {"W": 3, "D": 1, "L": 0}
merged_df["points"] = merged_df["result"].map(points_map)



# -------------------
# Rest days (difference in days between matches)
# -------------------
merged_df["datetime"] = pd.to_datetime(merged_df["datetime"], errors='coerce')

merged_df = merged_df.sort_values(["opponent", "datetime"]).reset_index(drop=True)

# Opponent goals average (goals conceded by this team)
merged_df["opponent_away_goals_avg"] = (
    merged_df.groupby("group_id")["goals_conceded"]
    .expanding()
    .mean()
    .reset_index(level=0, drop=True)
)
merged_df["opponent_avg_conceded_last5"] = (
    merged_df.groupby("opponent")["goals_conceded"]
    .transform(lambda x: x.shift().rolling(5, min_periods=1).mean())
)
merged_df["opponent_avg_goals_last5"] = (
    merged_df.groupby("opponent")["goals_scored"]
    .transform(lambda x: x.shift().rolling(5, min_periods=1).mean())
)
merged_df = merged_df.sort_values(["team", "datetime"]).reset_index(drop=True)

# Drop helper column
merged_df = merged_df.drop(columns=["points"])
# merged_df = merged_df.sort_values(["teamid", "datetime"]).reset_index(drop=True)

# # Define flags
# merged_df["is_win"]  = np.where(merged_df["result"] == "W", 1, 0)
# merged_df["is_draw"] = np.where(merged_df["result"] == "D", 1, 0)
# merged_df["is_loss"] = np.where(merged_df["result"] == "L", 1, 0)

# # Use transform instead of apply
# merged_df["last5matcheswins"] = (
#     merged_df.groupby("teamid")["is_win"]
#     .transform(lambda x: x.shift().rolling(5, min_periods=1).sum())
# )

# merged_df["last5matchesdraws"] = (
#     merged_df.groupby("teamid")["is_draw"]
#     .transform(lambda x: x.shift().rolling(5, min_periods=1).sum())
# )

# merged_df["last5matchesloss"] = (
#     merged_df.groupby("teamid")["is_loss"]
#     .transform(lambda x: x.shift().rolling(5, min_periods=1).sum())
# )
merged_df = merged_df.drop(columns=["is_win", "is_draw", "is_loss"])
```

```python
# we are going to predict goals scored and conceded using other features then with the predicted goals we will predict result
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Splitting the data for goals scored prediction
X_goals = merged_df.drop(columns=['goals_scored', 'goals_conceded'])
y_goals_scored = merged_df['goals_scored']
y_goals_conceded = merged_df['goals_conceded']
X_train_gs, X_test_gs, y_train_gs, y_test_gs = train_test_split(X_goals, y_goals_scored, test_size=0.2, random_state=42)
X_train_gc, X_test_gc, y_train_gc, y_test_gc = train_test_split(X_goals, y_goals_conceded, test_size=0.2, random_state=42)
# Model for goals scored
rf_gs = RandomForestRegressor(n_estimators=100, random_state=42)
rf_gs.fit(X_train_gs, y_train_gs)
y_pred_gs = rf_gs.predict(X_test_gs)
mae_gs = mean_absolute_error(y_test_gs, y_pred_gs)
mse_gs = mean_squared_error(y_test_gs, y_pred_gs)
rmse_gs = np.sqrt(mse_gs)
print(f"Goals Scored - MAE: {mae_gs}, MSE: {mse_gs}, RMSE: {rmse_gs}")
# Model for goals conceded
rf_gc = RandomForestRegressor(n_estimators=100, random_state=42)
rf_gc.fit(X_train_gc, y_train_gc)
y_pred_gc = rf_gc.predict(X_test_gc)
mae_gc = mean_absolute_error(y_test_gc, y_pred_gc)
mse_gc = mean_squared_error(y_test_gc, y_pred_gc)
rmse_gc = np.sqrt(mse_gc)
print(f"Goals Conceded - MAE: {mae_gc}, MSE: {mse_gc}, RMSE: {rmse_gc}")

```

```python
#hyperparameter tuning using gridsearchcv and randomizedsearchcv for goals scored model
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
rf = RandomForestRegressor(random_state=42)

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(X_train_gs, y_train_gs)
print(f"Best parameters from GridSearchCV: {grid_search.best_params_}")
best_rf_gs = grid_search.best_estimator_
y_pred_gs_best = best_rf_gs.predict(X_test_gs)
mae_gs_best = mean_absolute_error(y_test_gs, y_pred_gs_best)
mse_gs_best = mean_squared_error(y_test_gs, y_pred_gs_best)
rmse_gs_best = np.sqrt(mse_gs_best) 
print(f"Optimized Goals Scored - MAE: {mae_gs_best}, MSE: {mse_gs_best}, RMSE: {rmse_gs_best}")
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, n_iter=10, cv=3, n_jobs=-1, verbose=2, random_state=42)
random_search.fit(X_train_gs, y_train_gs)
print(f"Best parameters from RandomizedSearchCV: {random_search.best_params_}")
best_rf_gs_rand = random_search.best_estimator_
y_pred_gs_rand = best_rf_gs_rand.predict(X_test_gs)
mae_gs_rand = mean_absolute_error(y_test_gs, y_pred_gs_rand)
mse_gs_rand = mean_squared_error(y_test_gs, y_pred_gs_rand)
rmse_gs_rand = np.sqrt(mse_gs_rand)
print(f"Optimized (Random) Goals Scored - MAE: {mae_gs_rand}, MSE: {mse_gs_rand}, RMSE: {rmse_gs_rand}")
```

```python
#hyperparameter tuning using gridsearchcv and randomizedsearchcv for goals conceded model
grid_search_gc = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search_gc.fit(X_train_gc, y_train_gc)
print(f"Best parameters from GridSearchCV: {grid_search_gc.best_params_}")
best_rf_gc = grid_search_gc.best_estimator_
y_pred_gc_best = best_rf_gc.predict(X_test_gc)
mae_gc_best = mean_absolute_error(y_test_gc, y_pred_gc_best)
mse_gc_best = mean_squared_error(y_test_gc, y_pred_gc_best)
rmse_gc_best = np.sqrt(mse_gc_best)
print(f"Optimized Goals Conceded - MAE: {mae_gc_best}, MSE: {mse_gc_best}, RMSE: {rmse_gc_best}")
random_search_gc = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, n_iter=10, cv=3, n_jobs=-1, verbose=2, random_state=42)
random_search_gc.fit(X_train_gc, y_train_gc)
print(f"Best parameters from RandomizedSearchCV: {random_search_gc.best_params_}")
best_rf_gc_rand = random_search_gc.best_estimator_
y_pred_gc_rand = best_rf_gc_rand.predict(X_test_gc)
mae_gc_rand = mean_absolute_error(y_test_gc, y_pred_gc_rand)
mse_gc_rand = mean_squared_error(y_test_gc, y_pred_gc_rand)
rmse_gc_rand = np.sqrt(mse_gc_rand)
print(f"Optimized (Random) Goals Conceded - MAE: {mae_gc_rand}, MSE: {mse_gc_rand}, RMSE: {rmse_gc_rand}")
```
