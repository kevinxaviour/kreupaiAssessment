```python
#Collecting Matchids for all matches in a season
whoscored_url = 'https://1xbet.whoscored.com/regions/206/tournaments/4/seasons/9682/stages/22176/fixtures/spain-laliga-2023-2024'
driver = webdriver.Chrome()
driver.get(whoscored_url)
wait = WebDriverWait(driver, 10)
matchids2324 = []
while True:
    time.sleep(2)  
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Collecting all hrefs
    elements = soup.find_all(class_="Match-module_score__5Ghhj")
    hrefs = [tag['href'] for tag in elements if tag.has_attr('href')]

    # Extract match IDs (2nd or 3rd segment)
    for h in hrefs:
        parts = h.split("/")
        if len(parts) > 2:
            matchids2324.append(parts[2])
        elif len(parts) > 1:
            matchids2324.append(parts[1])

    print(f"Collected so far: {matchids2324}")

    # Try to click the button
    try:
        btn = wait.until(EC.element_to_be_clickable((By.ID, "dayChangeBtn-prev")))
        btn.click()
    except (TimeoutException, ElementClickInterceptedException, NoSuchElementException):
        print("Button is no longer clickable. Stopping.")
        break



driver.quit()

print("Final match IDs:", matchids2324)
```

```python
#scraping function to get event and player stats data
def extract_json_from_html(html_content, save_output=False):
    html = html_content

    regex_pattern = r'(?<=require\.config\.params\["args"\].=.)[\s\S]*?;'
    data_txt = re.findall(regex_pattern, html)
    
    if not data_txt:
        raise ValueError("No JSON data found with the specified pattern.")

    data_txt = data_txt[0]

    # Add quotations for JSON parser
    data_txt = data_txt.replace('matchId', '"matchId"')
    data_txt = data_txt.replace('matchCentreData', '"matchCentreData"')
    data_txt = data_txt.replace('matchCentreEventTypeJson', '"matchCentreEventTypeJson"')
    data_txt = data_txt.replace('formationIdNameMappings', '"formationIdNameMappings"')
    data_txt = data_txt.replace('};', '}')

    if save_output:
        # Save JSON data to txt
        output_file = open("output_data.txt", "wt", encoding='utf-8')
        output_file.write(data_txt)
        output_file.close()

    return data_txt

def extract_data_from_dict(data):
    # Load data from json
    event_types_json = data["matchCentreEventTypeJson"]
    formation_mappings = data["formationIdNameMappings"]
    events_dict = data["matchCentreData"]["events"]
    teams_dict = {data["matchCentreData"]['home']['teamId']: data["matchCentreData"]['home']['name'],
                  data["matchCentreData"]['away']['teamId']: data["matchCentreData"]['away']['name']}
    players_dict = data["matchCentreData"]["playerIdNameDictionary"]
    
    # Create players dataframe
    players_home_df = pd.DataFrame(data["matchCentreData"]['home']['players'])
    players_home_df["teamId"] = data["matchCentreData"]['home']['teamId']
    players_away_df = pd.DataFrame(data["matchCentreData"]['away']['players'])
    players_away_df["teamId"] = data["matchCentreData"]['away']['teamId']
    
    players_df = pd.concat([players_home_df, players_away_df])
    players_ids = data["matchCentreData"]["playerIdNameDictionary"]
    
    return events_dict, players_df, teams_dict

i=0
for matchid in df11:
    driver=webdriver.Chrome()
    url=f"http://1xbet.whoscored.com/matches/{matchid}/live/spain-laliga-2023-2024-girona-granada"
    driver.get(url)
    soup=BeautifulSoup(driver.page_source,'html.parser')
    match_html_content = str(soup)


    json_data_txt = extract_json_from_html(match_html_content)
    data = json.loads(json_data_txt)
    events_dict, players_df, teams_dict = extract_data_from_dict(data)
    
    df = pd.DataFrame(events_dict)
    dfp = pd.DataFrame(players_df)
    
    df['type'] = df['type'].apply(lambda x: x['displayName'])
    df['outcomeType'] = df['outcomeType'].apply(lambda x: x['displayName'])
    df['period'] = df['period'].apply(lambda x: x['displayName'])
    
    
    df['period'] = df['period'].replace({'FirstHalf': 1, 'SecondHalf': 2, 'FirstPeriodOfExtraTime': 3, 'SecondPeriodOfExtraTime': 4, 
                              'PenaltyShootout': 5, 'PostGame': 14, 'PreMatch': 16})
    
    df['teamName'] = df['teamId'].map(teams_dict)
    team_names = list(teams_dict.values())
    opposition_dict = {team_names[i]: team_names[1-i] for i in range(len(team_names))}
    df['oppositionTeamName'] = df['teamName'].map(opposition_dict)
    columns_to_drop = ['height', 'weight', 'age', 'isManOfTheMatch', 'field', 'stats', 
                       'subbedInPlayerId', 'subbedOutPeriod', 
                       'subbedOutExpandedMinute', 'subbedInPeriod', 'subbedInExpandedMinute', 'subbedOutPlayerId', 
                       'teamId']
    dfp.drop(columns=columns_to_drop, inplace=True)
    
    
    df = df.merge(dfp, on='playerId', how='left')
    
    df['name'] = df['name'].astype(str)
    df['name'] = df['name'].apply(unidecode)
    hteamID = list(teams_dict.keys())[0]  
    ateamID = list(teams_dict.keys())[1]  
    hteamName= teams_dict[hteamID]
    ateamName= teams_dict[ateamID]
    df['ishome'] = np.where(df['teamName'] == hteamName, 1, 0)
    element = soup.select_one('script:-soup-contains("matchCentreData")')
    json.loads(element.text.split("matchCentreData: ")[1].split(',\n')[0])
    matchdict = json.loads(element.text.split("matchCentreData: ")[1].split(',\n')[0])
    matchdict.keys()
    matchdict['timeStamp']
    df['datetime']=matchdict['timeStamp']
    
    df.to_csv(f"E:/KreupAI/DAta/Event/24-25/{hteamName} vs {ateamName}.csv")
    url=f"https://1xbet.whoscored.com/statisticsfeed/1/getmatchcentreplayerstatistics?category=summary&subcategory=all&statsAccumulationType=0&isCurrent=true&playerId=&teamIds={hteamID}&matchId={matchid}&stageId=&tournamentOptions=&sortBy=&sortAscending=&age=&ageComparisonType=&appearances=&appearancesComparisonType=&field=&nationality=&positionOptions=&timeOfTheGameEnd=&timeOfTheGameStart=&isMinApp=&page=&includeZeroValues=&numberOfPlayersToPick=&incPens="
    driver = webdriver.Chrome()
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    raw_content = soup.body.get_text(strip=True)
    datah = json.loads(raw_content)
    datah=datah['playerTableStats']
    datah=pd.DataFrame(datah)
    datah['teamname']=hteamName
    url=f"https://1xbet.whoscored.com/statisticsfeed/1/getmatchcentreplayerstatistics?category=summary&subcategory=all&statsAccumulationType=0&isCurrent=true&playerId=&teamIds={ateamID}&matchId={matchid}&stageId=&tournamentOptions=&sortBy=&sortAscending=&age=&ageComparisonType=&appearances=&appearancesComparisonType=&field=&nationality=&positionOptions=&timeOfTheGameEnd=&timeOfTheGameStart=&isMinApp=&page=&includeZeroValues=&numberOfPlayersToPick=&incPens="
    driver = webdriver.Chrome()
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    raw_content = soup.body.get_text(strip=True)
    dataa = json.loads(raw_content)
    dataa=dataa['playerTableStats']
    dataa=pd.DataFrame(dataa)
    dataa['teamname']=ateamName
    df = pd.concat([datah, dataa], ignore_index=True)
    df['ishome'] = np.where(df['teamname'] == hteamName, 1, 0)
    df['name'] = df['name'].apply(unidecode)
    df['datetime']=matchdict['timeStamp']
    df.to_csv(f'E:/KreupAI/DAta/PlayerStat/24-25/{hteamName} vs {ateamName}.csv')
    i+=1
    if i%15==0:
        driver.quit()
        time.sleep(120)

    

```
